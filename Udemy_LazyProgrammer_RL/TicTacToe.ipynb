```python
import numpy as np
import matplotlib.pyplot as plt
LENGTH = 3
```

```python
class Environment:
    
    def __init__(self):
        self.board = numpy.zeros(shape=(LENGTH,LENGTH))
        self.p1 = 1
        self.p2 = 2
        self.ended = False
        self.winner = None
        self.total_states = 3**(LENGTH*LENGTH)
        
    def game_over(self):
        #checking rows
        for i in range(LENGTH):
            for player in (self.p1, self.p2):
                if self.board[i].sum == player*LENGTH:
                    self.winner = player
                    self.ended = True
                    return True
        
        #checking columns
        for j in range(LENGTH):
            for player in (self.p1, self.p2):
                if self.board[:,j].sum == player*LENGTH:
                    self.winner = player
                    self.ended = True
                    return True
        
        #checking diagonals
        for player in (self.p1, self.p2):
            if self.board.trace() == player*LENGTH:
                self.winner = player
                self.ended = True
                return True
            
        for player in (self.p1, self.p2):
            if np.fliplr(self.board).trace() == player*LENGTH:
                self.winner = player
                self.ended = True
                return True
        
        #checking if draw
        if np.all((self.board) == 0) == False:
            self.ended = True
            self.winner = None
            return True
        
        #game not over yet
        self.winner = None
        self.ended = False
        return False

    def draw_board(self):
        for i in range(LENGTH):
            print("-------------")
            for j in range(LENGTH):
                print("  ", end="")
                if self.board[i,j] == self.p1:
                    print("1 ", end="")
                elif self.board[i,j] == self.p2:
                    print("2 ", end="")
                else:
                    print("  ", end="")
            print("")
        print("-------------")
        
```

```python
class Human:

    def play_game(p1, p2, env, draw=False):
        # loops until the game is over
        current_player = None
        while not env.game_over():
        # alternate between players
        # p1 always starts first
            if current_player == p1:
                current_player = p2
            else:
                current_player = p1

            # draw the board before the user who wants to see it makes a move
            if draw:
                if draw == 1 and current_player == p1:
                    env.draw_board()
                if draw == 2 and current_player == p2:
                    env.draw_board()

            # current player makes a move
            current_player.take_action(env)

            # update state histories
            state = env.get_state()
            p1.update_state_history(state)
            p2.update_state_history(state)

        if draw:
            env.draw_board()

        # do the value function update
        p1.update(env)
        p2.update(env)
```
